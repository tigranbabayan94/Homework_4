```python
import urllib2 as ul
from lxml import html
from lxml.html import document_fromstring
import json
import unicodecsv as csv
from pprint import pprint
url = "https://www.youtube.com/feed/trending"
response = ul.urlopen(url)
page = response.read()
tree = document_fromstring(page)
```

```python
UL=tree.xpath('//ul[@class="expanded-shelf-content-list has-multiple-items"]')
trending_all=[]
for j in range(len(UL)):
    x=UL[j].cssselect('h3')
    y=UL[j].cssselect('[class="yt-lockup-byline "]')
    z=UL[j].cssselect('[class="yt-lockup-meta-info"]')
    for i in range(len(x)):
        trending_content={}
        title=x[i].cssselect('a')[0].text_content()
        link=x[i].cssselect('a')[0].get('href')
        duration=str(x[i].cssselect('span')[0].text_content())
        user=y[i].cssselect('a')[0].text_content()
        view=z[i].cssselect('li')[1].text_content()
        trending_content["Title"] = title
        trending_content["Link"] = "https://www.youtube.com"+str(link)
        trending_content["Duration"] = duration[12:-1]
        trending_content["User"] = user
        trending_content["view"] = view
        trending_all.append(trending_content)
pprint(trending_all)
```

```python
with open('YouTube_url.json', 'w') as outcome:
    json.dump(trending_all, outcome, sort_keys = True, indent = 4)
```

```python
with open('YouTube_url.csv', 'w') as csv_file:
    for i in trending_all:
            writer = csv.writer(csv_file)
            for key, value in i.items():
                writer.writerow([key, value])
            writer.writerow([])
```
