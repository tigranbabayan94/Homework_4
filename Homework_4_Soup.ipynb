```python
from BeautifulSoup import *
import requests
from pprint import pprint
import json
import unicodecsv as csv
url="https://www.youtube.com/feed/trending"
response = requests.get(url)
data = response.text
soup = BeautifulSoup(data)
```

```python
title_url = soup.findAll('a', attrs={'class': 'yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink      spf-link '})
duration = soup.findAll('span', attrs={'class': 'accessible-description'})
user = soup.findAll('a', attrs={'class': 'g-hovercard yt-uix-sessionlink       spf-link '})
view = soup.findAll("ul", attrs={ 'class':"yt-lockup-meta-info", } )
```

```python
view1=[]
for i in view:
    for j in i:
        view1.append(j.text)
view2=view1[1::2]
dictionary = []
```

```python
for w,x,y,z in zip(title_url, duration, user, view2):
    dictionary.append({'Title ' :w.get("title")})
    dictionary.append({"URL" : "https://www.youtube.com"+w.get('href')})
    dictionary.append({'Duration':x.text})
    dictionary.append({"Username":y.text})
    dictionary.append({"Views": z})
```

```python
pprint(dictionary)
```

```python
with open('YouTube.json', 'w') as trending:
    json.dump(dictionary, trending, sort_keys = True, indent = 4)
```

```python
with open('YouTube.csv', 'w') as trending1:
    for i in dictionary:
            writer = csv.writer(trending1)
            for i, x in i.items():
                writer.writerow([i, x])
            writer.writerow([])
```
